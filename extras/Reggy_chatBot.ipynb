```python
import re

def chatbot():
    # Simple state management
    user_data = {"name": None, "mood": None}

    print("Reggy: Hi! I'm a pattern-matching bot. What's your name?")

    while True:
        user_input = input("You: ").strip()

        # 1. EXIT CONDITION
        if re.search(r'\b(exit|quit|bye)\b', user_input.lower()):
            print(f"Reggy: Goodbye {user_data['name'] or 'friend'}!")
            break

        # 2. CAPTURING NAME (Looking for "My name is [Name]" or just "[Name]")
        if not user_data["name"]:
            # This regex looks for 'name is X' OR captures the first word if it's a simple name
            match = re.search(r"(?:name is|i am|i'm)\s+([a-zA-Z]+)", user_input, re.I)
            if match:
                user_data["name"] = match.group(1)
                print(f"Reggy: Nice to meet you, {user_data['name']}! How are you feeling today?")
                continue
            else:
                # Fallback if they just typed their name as one word
                user_data["name"] = user_input.split()[0]
                print(f"Reggy: Thanks, {user_data['name']}! How are you feeling today?")
                continue

        # 3. ANALYZING MOOD (Using regex keywords)
        if not user_data["mood"]:
            if re.search(r"\b(good|happy|great|well|ok)\b", user_input.lower()):
                user_data["mood"] = "good"
                print(f"Reggy: I'm glad to hear that, {user_data['name']}! What's on your mind?")
            elif re.search(r"\b(bad|sad|tired|angry|stressed)\b", user_input.lower()):
                user_data["mood"] = "bad"
                print(f"Reggy: I'm sorry you're feeling {user_input}. Want to talk about it?")
            else:
                print("Reggy: I see. Tell me more about that.")
            continue

        # 4. GENERAL CONVERSATION (Reflecting user input)
        # Using a capture group to flip "I am" to "Why are you"
        match = re.search(r"i am (.*)|i'm (.*)", user_input.lower())
        if match:
            content = match.group(1) or match.group(2)
            print(f"Reggy: Why are you {content}?")
        else:
            print("Reggy: That's interesting. Tell me more!")

chatbot()
```
