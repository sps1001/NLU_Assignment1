```python
# RollNumber_prob1.py
import re
from datetime import date

def calculate_age(day, month, year):
    today = date.today()
    age = today.year - year
    if (today.month, today.day) < (month, day):
        age -= 1
    return age

def parse_birthday(text):
    text = text.strip()

    # Format: dd-mm-yyyy or dd-mm-yy
    match = re.search(r"\b(\d{1,2})[-/](\d{1,2})[-/](\d{2,4})\b", text)
    if match:
        d, m, y = match.groups()
        d, m, y = int(d), int(m), int(y)
        if y < 100:
            y += 2000 if y < 25 else 1900
        return d, m, y

    # Format: dd Month yyyy
    match = re.search(r"\b(\d{1,2})\s+([A-Za-z]+)\s+(\d{4})\b", text)
    if match:
        d, month_name, y = match.groups()
        months = {
            "january":1,"jan":1,"february":2,"feb":2,"march":3,"mar":3,
            "april":4,"apr":4,"may":5,"june":6,"jun":6,"july":7,"jul":7,
            "august":8,"aug":8,"september":9,"sep":9,"october":10,"oct":10,
            "november":11,"nov":11,"december":12,"dec":12
        }
        month = months.get(month_name.lower())
        if month:
            return int(d), month, int(y)

    return None

def detect_mood(text):
    text = text.lower()

    if re.search(r"\b(happy|hapy|good|gud|great|grate|awesome|fine|ok|okay)\b", text):
        return "good"
    if re.search(r"\b(sad|saad|bad|tired|tyred|angry|angri|stressed|stressd)\b", text):
        return "bad"
    return "neutral"

def chatbot():
    user_data = {"name": None, "surname": None, "age": None}

    print("Reggy: Hi! What is your full name?")

    while True:
        user_input = input("You: ").strip()

        # Exit
        if re.search(r"\b(exit|quit|bye)\b", user_input.lower()):
            print(f"Reggy: Goodbye {user_data['name'] or 'friend'}!")
            break

        # Capture full name
        if not user_data["name"]:
            parts = user_input.split()
            user_data["name"] = parts[0]
            if len(parts) > 1:
                user_data["surname"] = parts[-1]
                print(f"Reggy: Nice to meet you {user_data['name']} {user_data['surname']}!")
            else:
                print(f"Reggy: Nice to meet you {user_data['name']}!")
            print("Reggy: When is your birthday?")
            continue

        # Capture birthday
        if not user_data["age"]:
            parsed = parse_birthday(user_input)
            if parsed:
                d, m, y = parsed
                age = calculate_age(d, m, y)
                user_data["age"] = age
                print(f"Reggy: You are {age} years old!")
                print("Reggy: How are you feeling today?")
            else:
                print("Reggy: I couldn't understand that date. Please try again.")
            continue

        # Capture mood
        mood = detect_mood(user_input)
        if mood == "good":
            print("Reggy: That's wonderful! Keep smiling!")
        elif mood == "bad":
            print("Reggy: I'm sorry to hear that. Things will get better!")
        else:
            print("Reggy: I see. Tell me more about it.")

chatbot()
```
